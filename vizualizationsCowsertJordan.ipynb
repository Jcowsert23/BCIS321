```python
#visualization 1
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

shots = pd.read_csv('https://raw.githubusercontent.com/jcowsert23/BCIS321/main/shot_log-Jordan_Cowsert.csv')

plt.hist(shots.TOUCH_TIME, bins=25, color = "pink")
plt.show()
```

```python
#visualization 2
import seaborn as sns

sns.boxplot(shots['SHOT_CLOCK'],color = 'teal', orient= 'h')
plt.title('Box Plot for Shot Clock Times')
plt.xlabel('Shot Clock')
plt.show()
```

```python
#Vizualization 3
sns.countplot(x=shots['SHOT_RESULT'])
plt.title("Shot Results Graph", fontsize=18)
plt.xlabel("Shot Result", fontsize=18)
plt.ylabel('Counts', fontsize=18)
```
